```python


import numpy as np

def gradient_descent(x, y, learning_rate=0.01, iterations=1000):




    n = len(x)
    m = 0  # Initialize slope
    b = 0  # Initialize intercept

    for _ in range(iterations):
        y_predicted = m * x + b
        dm = -(2/n) * sum(x * (y - y_predicted))
        db = -(2/n) * sum(y - y_predicted)
        m = m - learning_rate * dm
        b = b - learning_rate * db

    return m, b

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Perform gradient descent with different learning rates
learning_rates = [0.001, 0.01, 0.1]

for lr in learning_rates:
    m, b = gradient_descent(x, y, learning_rate=lr)
    print(f"Learning Rate: {lr}, Slope (m): {m}, Intercept (b): {b}")
    y_pred = m * x + b
    print(f"Predictions: {y_pred}") #Print predictions for the given learning rate
```
